```python

import json

filename = "D:/rdr_dataset_12/all_captures.json"

with open(filename, "r", encoding="utf-8") as f:
    content = f.read()

decoder = json.JSONDecoder()

# Parse the first JSON object
_, end = decoder.raw_decode(content)

# Remove the first object and any whitespace following it
remaining = content[end:].lstrip()

# Overwrite the file with the remaining content
with open(filename, "w", encoding="utf-8") as f:
    f.write(remaining)
```

```python
# read csv
import csv
import json

# new json object
data = []

with open('weather_annotation_subset_orig.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        choice = row[2]
        if choice.startswith('{"choices":'):
            choices = json.loads(choice)['choices']

            
        else:
            choices = [choice]

        # multi-hot encode the choices ['clear', 'cloudy', 'overcast', 'fog', 'rain', 'snow']
        choices = [1 if c in choices else 0 for c in ['clear', 'cloudy', 'overcast', 'fog', 'rain', 'snow']]

        print(choices)
        data.append({"image": row[5], "labels": choices})
        #print(row)
print(data)

# write to json
with open('weather_annotation_subset_path_choicemh.json', 'w') as f:
    json.dump(data, f)
```

```python
from datasets import load_dataset

dataset = load_dataset("imagefolder", data_dir="D:/WeatherDataset_pngs")
```

```python
from transformers import AutoImageProcessor, AutoModelForImageClassification

processor = AutoImageProcessor.from_pretrained("./outputs")
model = AutoModelForImageClassification.from_pretrained("./outputs/checkpoint-3000")

model.eval()
```

```python
import json
import tqdm
from PIL import Image
import torch


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
with open("G:/rdr_dataset_12/all_captures.json", 'r+') as json_file:
    dict_all_captures = json.load(json_file)
    ids = dict_all_captures.keys()
    for id in tqdm.tqdm(ids):
        image = Image.open("G:/rdr_dataset_12/png_12/o_" + id + "_12.png").convert("RGB")

        inputs = processor(images=image, return_tensors="pt")

        
        inputs.to(device)
        with torch.no_grad():   
            outputs = model(**inputs)
        
        logits = outputs.logits
        
        predicted_class_idx = logits.argmax(-1).item()

        label = model.config.id2label[predicted_class_idx]

        dict_all_captures[id]["Capture"]["Weather_Predicted"] = label
    
    json.dump(dict_all_captures, json_file)

    
```
